# Recolección de tiempos de ejecución con diferentes tamaños de matrices Cholesky

Importes necesarios

```python
import matplotlib.pyplot as plt
import numpy as np
```

## C

Recolectamos los resultados para el algoritmo de C sin paralelismo (Cholesky)

```python
#matrices de tamaños: 10, 100, 200, 300, 400, 500
tamaños = [10, 100, 200, 300, 400, 500]
```

```python
resultados_c = [
    [0.0057, 0.0061, 0.0057, 0.0053, 0.0031],
    [0.45, 0.4295, 0.4577, 0.4432, 0.4479],
    [3.4321, 4.6542, 3.4813, 4.6957, 4.3309],
    [13.2008, 12.8538, 12.8921, 12.7065, 12.7327],
    [30.2981, 31.7176, 31.4273, 31.7424, 31.8156],
    [61.6718, 62.8253, 62.4126, 61.0508, 62.3506]
]
```

```python
resultados_c_paralelo = [[0.007331, 0.005269, 0.004162, 0.003006, 0.004810],
                         [0.003198, 0.002997, 0.003168, 0.006641, 0.003060],
                         [0.013222, 0.009616, 0.008802, 0.008923, 0.009470],
                         [0.026964, 0.030298, 0.026056, 0.030188, 0.024914],
                         [0.054207, 0.053929, 0.054187, 0.055820, 0.063224],
                         [0.103280, 0.106633, 0.107698, 0.116829, 0.103316]]
```

```python
promedios_c = [np.mean(r) for r in resultados_c]
desviaciones_c = [np.std(r) for r in resultados_c]
promedios_c_paralelo = [np.mean(r) for r in resultados_c_paralelo]
desviaciones_c_paralelo = [np.std(r) for r in resultados_c_paralelo]

promedios_c = np.array(promedios_c)
desviaciones_c = np.array(desviaciones_c)
promedios_c_paralelo = np.array(promedios_c_paralelo)
desviaciones_c_paralelo = np.array(desviaciones_c_paralelo)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(tamaños, promedios_c, color='blue', label='Promedio')

plt.fill_between(tamaños, promedios_c - desviaciones_c, promedios_c + desviaciones_c,
                 color='blue', alpha=0.2, label='± Desviación estándar')

plt.title('Tiempo de ejecución')
plt.xlabel('Tamaño de matriz')
plt.ylabel('Tiempo de inversión (segundos)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(tamaños, promedios_c_paralelo, color='blue', label='Promedio')

plt.fill_between(tamaños, promedios_c_paralelo - desviaciones_c_paralelo, promedios_c_paralelo + desviaciones_c_paralelo,
                 color='blue', alpha=0.2, label='± Desviación estándar')

plt.title('Tiempo de ejecución')
plt.xlabel('Tamaño de matriz')
plt.ylabel('Tiempo de inversión (segundos)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
