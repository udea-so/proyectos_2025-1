# Recolección de tiempos de ejecución con diferentes tamaños de matrices LU

## Python

Calculamos el promedio y la desviación estándar

```python
tamaños = np.array(tamaños)

promedios = [np.mean(r) for r in resultados_python]
desviaciones = [np.std(r) for r in resultados_python]
promedios_paralelo = [np.mean(r) for r in resultados_python_paralelo]
desviaciones_paralelo = [np.std(r) for r in resultados_python_paralelo]

promedios = np.array(promedios)
desviaciones = np.array(desviaciones)
promedios_paralelo = np.array(promedios_paralelo)
desviaciones_paralelo = np.array(desviaciones_paralelo)
```

Graficamos la media de tiempos de inversión no paralelizada con respecto a las dimensiones de la matriz.

```python
plt.figure(figsize=(8, 5))
plt.plot(tamaños, promedios, color='blue', label='Promedio')

plt.fill_between(tamaños, promedios - desviaciones, promedios + desviaciones,
                 color='blue', alpha=0.2, label='± Desviación estándar')

plt.title('Tiempo de ejecución')
plt.xlabel('Tamaño de matriz')
plt.ylabel('Tiempo de inversión (segundos)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(tamaños, promedios_paralelo, color='blue', label='Promedio')

plt.fill_between(tamaños, promedios_paralelo - desviaciones_paralelo, promedios_paralelo + desviaciones_paralelo,
                 color='blue', alpha=0.2, label='± Desviación estándar')

plt.title('Tiempo de ejecución')
plt.xlabel('Tamaño de matriz')
plt.ylabel('Tiempo de inversión (segundos)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
